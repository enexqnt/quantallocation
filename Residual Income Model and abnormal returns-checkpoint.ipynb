### Equity Risk Premium

```python
## import historical equity risk premium calculated by Damodaran
erp=pd.read_excel('erp_damodaran.xlsx', index_col=0, parse_dates=True).dropna()
erp.index = pd.to_datetime(erp.index)
erp=erp.resample('M').ffill()
erp+=0.02
erp.plot(figsize=(15,5),title='Equity Risk Premium')

plt.savefig('rim_erp.png')
```

### Cost of equity

```python
## create a common data-frame with the same index
a=pd.merge_asof(erp, rf, on="Date", direction='nearest')
a=pd.merge_asof(a, beta, on="Date", direction='nearest')
```

```python
## set the index equal to the dates
a=a.dropna().set_index('Date')
```

```python
## calculate Market Risk Premium - Risk Free Rate
a['MRP-RF']=a['ERP']-a['IT YIELD 10-YEAR GOVERNMENT BONDS NADJ']
```

```python
## (MRP - RF) X ADJ_BETA
s=a.iloc[:,2:].multiply(a['MRP-RF'], axis='index') 
```

```python
## Calculate cost of equity adding the risk free rate
ke=s.add(a['IT YIELD 10-YEAR GOVERNMENT BONDS NADJ'],axis=0) 
```

```python
ke.clip(lower=0.01,inplace=True)
ke.plot(legend=False,figsize=(15,5),title='Cost of Equity')
plt.savefig('rim_ke.png')
```

The forward BVPSy1 has been calculated through the â€˜â€˜clean surplus accountingâ€™â€™ relationship (BVPSy1 = BVPSy0+ EPSy1 â€“ DPSy1) using analystsâ€™ estimates on earnings and dividends.

```python
iv=((bvps+ (eps1-(ke*bvps))/(1+ke)+
     (eps2-(ke*(bvps+eps1-dvps1)))/((1+ke)**2))+ (eps3-(ke*(bvps+eps1-dvps1+eps2-dvps2)))/(((1+ke)**3)*(ke-g_m))
)
```

# 3. Backtest

```python
p.plot(legend=False, title='Historical prices',figsize=(15,5))
plt.savefig('rim_hist.png')
```

```python
iv_cleaned=p_iv[p_iv>0]
```

```python
wealth=(1+p.pct_change()).cumprod().dropna().loc[iv_cleaned.index[0]:]
wealth.plot(legend=False, title='Cumulative returns')
```

```python
def Rebalancing(X,frequency=4,quantile=0.1,worst=False):
    w=pd.DataFrame(0,columns=X.columns,index=X.index)
    for i in range(0,len(X)):
        if i%frequency==0:
            if worst==False:
                l=list(X.iloc[i][X.iloc[i]<X.iloc[i].quantile(quantile)].index)
            elif worst==True:
                l=list(X.iloc[i][X.iloc[i]>X.iloc[i].quantile(quantile)].index)
            n=len(l)
            ww=1/n
            w.loc[w[l].index[i],l]=[ww]*n
        else:
            w.iloc[i]=w.iloc[i-1]
    return w

def Wealth(w,prices):
    wealth1=(1+prices.pct_change()).cumprod().loc[w.index[0]:]
    ptf=(w*wealth1).dropna().sum(axis=1)
    return ptf

```

```python
w_top=Rebalancing(X=iv_cleaned,quantile=0.1,worst=False)

w_worst=Rebalancing(iv_cleaned,quantile=0.1,worst=True)

w_q=Rebalancing(iv_cleaned,quantile=0.4,worst=False)

equally_w=((1/len(iv_cleaned.columns))*wealth).sum(axis=1)
```

```python
total_returns=pd.read_excel(path,sheet_name='TR',index_col='Date')[validtickers].dropna().resample('MS').ffill()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(20, 10))

plt.plot(Wealth(w_top,total_returns),label='Top 10%')
plt.plot(Wealth(w_q,total_returns),label='Top 40%')


plt.plot(equally_w,label='EW')
plt.legend()
plt.show()
plt.savefig('rim_results.png')
```
